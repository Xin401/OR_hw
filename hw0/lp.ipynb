```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog
```

```python
#3-a
# 定義目標函數的係數
A = 1
B = -4
c = [-2, -A]

# 定義不等式約束條件的係數
A_ub = [[1, 1], [-2, 1], [0, 1], [-1, 0], [0, -1]]

# 定義不等式約束條件的右側值
b_ub = [8, -12, B, 0, 0]

# 使用 linprog 函數來解決線性規劃問題
res = linprog(c, A_ub=A_ub, b_ub=b_ub)

print('Optimal value:', res.fun)
print('Optimal solution:', res.x)

# 建立 x 的值
x = np.linspace(0, 10, 500)

# 繪製每個約束條件的邊界
plt.plot(x, 8 - x, label='x + y <= 8')
plt.plot(x, -12 + 2*x, label='-2x + y <= -12')
plt.plot(x, np.full_like(x, B), label='y <= -4')
plt.plot(np.full_like(x, 0), x, label='x >= 0')
plt.plot(x, np.full_like(x, 0), label='y >= 0')

# 繪製可行區域
y1 = np.maximum.reduce([8 - x, -12 + 2*x, np.full_like(x, B), np.full_like(x, 0)])
y2 = np.minimum.reduce([8 - x, -12 + 2*x, np.full_like(x, B), np.full_like(x, 0)])
plt.fill_between(x, y1, y2, where=(y1<y2), color='gray', alpha=0.5)

# 繪製最優解
# plt.plot(res.x[0], res.x[1], 'ro')

plt.xlim((0, 10))
plt.ylim((-10, 10))
plt.xlabel('x')
plt.ylabel('y')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
